### Task-3: Data Visualization with Matplotlib.

#### Description:
This task involves using the Matplotlib library to visualize data.
#### Responsibility:
Create a Scatter plot, Bar chart, Line chart, Histogram and Pie chart using Matplotlib to visualize data from a Pandas DataFrames.                      Customize the charts with labels, titles and legends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv("C:\\Users\\Manish Yadav\\Downloads\\householdtask3.csv")
```

```python
data.head()
```

```python
#Scatter plot with the year against own
plt.scatter(data['year'],data['own'], label='Ownership')

#Adding title to plot
plt.title('Scatter Plot')

#Setting the x and y labels
plt.xlabel('Year')
plt.ylabel('Own')

#Adding the legend
plt.legend()  #Automatically uses the label provided

#Showing the result
plt.show()
```

```python
#Line Chart with the year against own
plt.plot(data['year'],label='Year')
plt.plot(data['own'],label='Own')

#Adding title to Plot
plt.title('Line Chart')

#Setting the x and y labels
plt.xlabel('Year')
plt.ylabel('Own')

#Adding the legend
plt.legend()  #Automatically uses the label provided

#Showing the result
plt.show()
```

```python
#Bar Chart or Bar Plot
plt.bar(data['year'],data['own'],label='Ownership')

#Adding title to plot
plt.title('Bar Chart')

#Setting the x and y labels
plt.xlabel('Year')
plt.ylabel('Own')

#Adding the legend
plt.legend(loc='upper left', fontsize=7)  #Automatically uses the label provided

#Showing the result
plt.show()
```

```python
#Histogram
plt.hist(data['income'],label='Income')

#Title
plt.title('Histogram')

#Setting the x and y labels
plt.xlabel('Income')
plt.ylabel('People')

#Adding the legend
plt.legend()  #Automatically uses the label provided

#Showing the results
plt.show()
```

```python
data1 = data.head()

# Creating a pie chart
plt.pie(data1['income'],labels=data1['age'], autopct='%1.f%%', startangle=90)

# Adding title to the plot
plt.title('Pie Chart')

# Adding a legend
plt.legend(title="Age", loc="upper right", fontsize=5.6)  # loc adjusts legend position

# Showing the result
plt.show()
```
