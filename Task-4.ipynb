```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("C:\\Users\\Manish Yadav\\Downloads\\USvideos.csv")
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(['thumbnail_link','description'],axis=1,inplace=True)
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'],format='%y.%d.%m')  #format year,date and month properly
df.head(3)
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])  #format date and time properly in the publish_time column.
df.head(3)
```

```python
#Adding new columns fetching details from publish_time column
df['publish_month'] = df['publish_time'].dt.month
df['publish_day'] = df['publish_time'].dt.day
df['publish_hour'] = df['publish_time'].dt.hour
df.head(4)
```

```python
#Adding new column with values
df['category_name'] = df['category_id'].map({1:'Film and Animation',
    2:'Autos and Vehicles', 10:'Music',
    15:'Pets and Animals',
    17:'Sports', 19:'Travel and Events',
    20:'Gaming', 22:'People and Blogs',
    23:'Comedy', 24:'Entertainments',
    25:'News and Politics', 26:'How to and Style',
    27:'Education',28:'Science and Technology',
    29:'Non Provism and Activitism',
    30:'Movies', 43:'Shows'})
```

```python
df.head(5)
```

```python
df['year'] = df['publish_time'].dt.year
yearly_counts=df.groupby('year')['title'].count()

yearly_counts.plot(kind='bar',xlabel='Year',ylabel='Total Publish Count',title='Total Publish Video Per Year')
plt.show()
```

```python
plt.figure(figsize=(9,5))

category_view=df.groupby('category_name')['views'].sum().reset_index()

top_category=category_view.sort_values(by='views',ascending=False).head(5)

plt.bar(top_category['category_name'],top_category['views'])
plt.xlabel('Category Name',fontsize=12)
plt.ylabel('Total Views',fontsize=12)
plt.title('Top 5 Category Name',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='category_name',data=df,order=df['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Video Count by Category')
plt.show()
```

```python
videos_per_hour=df['publish_hour'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.barplot(x=videos_per_hour.index,y=videos_per_hour.values,palette='rocket')
plt.xticks()
plt.xlabel('Hour of Day')
plt.ylabel('Number of Videos')
plt.title('No of videos published per hour')
plt.show()
```

```python
df['publish_time']=pd.to_datetime(df['publish_time'])
df['publish_date']=df['publish_time'].dt.date
video_count_by_date=df.groupby('publish_date').size()

plt.figure(figsize=(12,6))
sns.lineplot(data=video_count_by_date)
plt.xticks(rotation=45)
plt.xlabel('Publish Date')
plt.ylabel('Number of Vidoes')
plt.title('Videos published per Time')
plt.show()
```

```python
sns.scatterplot(data=df, x='views', y='likes')
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplots_adjust(wspace=0.2,hspace=0.4,top=0.9)
plt.subplot(2,2,1)

g=sns.countplot(x='comments_disabled',data=df)
g.set_title('Comment Disable', fontsize=16)
plt.subplot(2,2,2)

g1=sns.countplot(x='ratings_disabled',data=df)
g1.set_title('Rating Disable', fontsize=16)
plt.subplot(2,2,3)

g2=sns.countplot(x='video_error_or_removed',data=df)
g2.set_title('Video Error or Removed', fontsize=16)
plt.show()
```
